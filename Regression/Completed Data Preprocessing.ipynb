# Data Preprocessing Tools

## Importing the libraries

```
#numpy helps you work with arrays
#pandas helps you work with dependent variable, matrix of features
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Data.csv')
#Matrix of features 
#iloc- locate indices
#Just put ":"- all rows
#:-1 all columns minus one
X = dataset.iloc[:,:-1].values
#Dependent variable
y = dataset.iloc[:,-1].values
```

```
print(X)
```

```
print(y)
```

## Taking care of missing data

```
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
#Dont include first since not numbers (only want numerical)
imputer.fit(X[:,1:3])
#transform returns fixed data
X[:,1:3] = imputer.transform(X[:,1:3])
```

```
print(X)
```

## Encoding categorical data

### Encoding the Independent Variable

```
#One hot encoding- country into 3 columns (e.g. isFrance, isSpain, isGermany)
#Don't want it to be one numerical variable as it makes it seem as order

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

#transformation- encoder, type of encoding- OneHot
#passthrough ensures we keep non categorical columns
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])], remainder='passthrough')

X = np.array(ct.fit_transform(X))
```

```
print(X)
```

### Encoding the Dependent Variable

```
from sklearn.preprocessing import LabelEncoder
#binary outcome
le = LabelEncoder()
y = le.fit_transform(y)
#one variable- still "ordering"
```

```
print(y)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```
print(X_train)
```

```
print(X_test)
```

```
print(y_train)
```

```
print(y_test)
```

## Feature Scaling

```
#Must be after train/test split- test set is ideally to act as a brand new set
#Would use some test data (information leakage) in calculating sample statistics if done before

#Scaling variables/features- all take values into same scale- makes coefficients more valuable

#Standardization x - u /sd (e.g. -3 to +3) works for all the time
#Normalization x - min / range (0 to 1) for normal distribution of features

#Don't scale dummy variables (already in same scale range as other variables), can confuse which country is which
```

```
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train[:,3:] = sc.fit_transform(X_train[:,3:])

#Must use same scaler for test set- don't fit
X_test[:,3:] = sc.transform(X_test[:,3:])
```

```
print(X_train)
```
